```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris_data=load_iris()     #iris_data is nothing but a dict
```

```python
iris_data
```

```python
print(iris_data['data'])
print(iris_data['target'])
```

```python
print("Feature names : ",iris_data.feature_names)
print("Target classes : ",iris_data['target_names'])
print(iris_data.keys())
```

```python
iris_data['data'].shape
```

```python
pd.Series(iris_data['target'])
```

```python
df=pd.DataFrame(iris_data['data'],columns=iris_data['feature_names'])
```

```python
df.head()
```

```python
df['target']=iris_data['target']
```

```python
df.head()
```

```python
names_target=[iris_data['target_names'][i] for i in iris_data['target']]
```

```python
names_target
```

```python
df['named_target']=names_target
```

```python
df.head()
```

```python
print('keys : ',df.keys())
```

### Creating classification model

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import train_test_split
```

```python
color=['red','green','pink']
#eg:
df[df['named_target']=='setosa']           #filtering the rows 
```

```python
plt.figure(figsize=(10,8))
for i in range(3):
    filtered_df=df[df['named_target']==iris_data['target_names'][i]]
    plt.scatter(filtered_df['petal length (cm)'],filtered_df['petal width (cm)'],c=color[i],label=iris_data['target_names'][i])
    
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,:4],df['named_target'],test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
predicted=model.predict(x_test)
predicted
```

```python
y_test
```

```python
model.score(x_test,y_test)   #96% accuracy
```

```python
sum(y_test==predicted)/len(y_test)
```

```python
plt.scatter(y_test,predicted)    #should be a straight line if there's 100 percent accuracy
```

```python
import seaborn as sn

confusion_matrix = pd.crosstab(y_test,predicted)

sn.heatmap(confusion_matrix, annot=True)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
